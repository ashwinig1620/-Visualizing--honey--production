```python
#1. How has honey production yield changed from 1998 to 2021?

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df1=pd.read_csv(r'C:\Users\Dell\Downloads\honeyproduction 1998-2021 (2).CSV')

df1
```

```python

average_yield = df1.groupby('year')['yieldpercol'].mean()
average_yield 
```

```python
plt.figure(figsize=(12,6),facecolor='orange')
average_yield.plot(kind='line',marker ='^',markersize=15,color='m')
plt.title('Honey Production Yield (1998-2021)',weight='bold',fontsize=20)
plt.xlabel('year',weight='bold',fontsize=18)
plt.ylabel('yield per colony',weight='bold',fontsize=18)
plt.xticks(range(1998,2022),rotation=45,color='maroon')
plt.yticks(range(40,80,5),color='blue')
plt.grid(True,color='green')
plt.show()
```

```python
# Over time, what are the major production trends across the states?
df1.head()
df1.info()
```

```python
r2 = df1.groupby('State')['totalprod'].sum().reset_index()
r2
```

```python
r1 = r2.sort_values(by='totalprod', ascending=True)
r1
```

```python
plt.figure(figsize=(18,12),facecolor='cyan')
plt.barh(r1['State'], r1['totalprod'],edgecolor='white')

for i, totalprod in enumerate(r1['totalprod']):
    plt.text(totalprod, i, str(totalprod), ha='left', va='center')
ax=plt.gca()
ax.set_facecolor('violet')
plt.xlabel('Total Production',weight='bold',fontsize=18,color='darkgreen')
plt.ylabel('state',weight='bold',fontsize=18,color='darkgreen')
plt.title('Honey Production by State',weight='bold',fontsize=20,color='magenta')
plt.xticks()
plt.yticks()
plt.show()
```

```python
# Does the data show any trends in terms of the number of honey producing colonies and yield per colony before 2006, which was when
#concern over Colony Collapse Disorder spread nationwide?
df1.head()
df1.info()
```

```python
filtered_data = df1[df1['year'] < 2006]
filtered_data
```

```python
grouped_data = filtered_data.groupby('year').agg({'numcol': 'sum', 'yieldpercol': 'mean'}).reset_index()
grouped_data
```

```python
plt.figure(figsize=(10, 6),facecolor='yellow')
plt.plot(grouped_data.index, grouped_data['numcol'], label='Number of Colonies')
plt.plot(grouped_data.index, grouped_data['yieldpercol'], label='Yield per Colony')
plt.xlabel('Year')
plt.ylabel('yield per colony')
plt.title('Trends in Honey Production')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plotting the trends
plt.figure(figsize=(10, 6),facecolor='yellow')
plt.bar(grouped_data['year'], grouped_data['numcol'],color='orange',edgecolor='black')
plt.bar(grouped_data['year'], grouped_data['yieldpercol'])
plt.xlabel(('Year'),weight='bold')
plt.ylabel(('yield per colony'),weight='bold')
plt.title('Trends in Honey Production',weight='bold')
plt.xticks(color='green')
plt.yticks(color='magenta')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df1=pd.read_csv(r'C:\Users\Dell\Downloads\honeyproduction 1998-2021 (2).CSV')

df1
```

```python
df_grouped = df1.groupby(df1['year'])['prodvalue'].sum()
df_grouped
```

```python
plt.figure(figsize=(12,5))
plt.plot(df_grouped.index, df_grouped.values,marker='o',color='blue')
ax = plt.gca()
ax.set_facecolor("grey")
plt.xlabel('Year',weight='bold',fontsize=18)
plt.ylabel('Production Value',weight='bold',fontsize=18)
plt.xticks(range(1998,2022),rotation=90,color='darkgreen')
plt.yticks(color='red')
plt.title('Changes in Production Value over the Years',weight='bold',fontsize=22)
plt.grid(True,color='black')
plt.show()
```
